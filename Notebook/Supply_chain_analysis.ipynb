#  MONTHLY SALES

```python
df['order_date'] = pd.to_datetime(df['order_date (DateOrders)'])

monthly_sales = df.groupby('order_date')['Sales'].sum().resample('ME').sum()
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.plot(monthly_sales)

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(monthly_sales, order=(5,1,0))

model_fit = model.fit()

forecast = model_fit.forecast(steps=3)

print(forecast)
```

```python
plt.figure(figsize=(10,5))

plt.plot(monthly_sales, label="Actual Sales")

plt.plot(forecast, label="Forecast", linestyle="dashed")

plt.legend()

plt.title("Supply Chain Demand Forecast")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.show()
```
